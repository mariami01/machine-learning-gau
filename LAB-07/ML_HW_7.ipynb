3. ააგეთ პლინომიალური რეგრესიული მრუდი, დააგენერირეთ შემთხვევითი მონაცემები (მინიმუმ 50) X[10; 20]
შაულედიდან, Y[500, 1000] შუალედიდან.

```python
import matplotlib.pyplot as plt
import numpy
import numpy as np
x = np.random.randint(low=10, high=20, size=50)
y = np.random.randint(low=500, high=1000, size=50)

mymodel = numpy.poly1d(numpy.polyfit(x, y, 2))
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.title('Polynomial Regression Line')
plt.xlabel('Y elemnts')
plt.ylabel('X elements')
plt.show()
```
